```python
column = df[["age", "campaign", "duration"]]
q1 = np.percentile(column, 25)
q3 = np.percentile(column, 75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df[["age", "campaign", "durattion"]] = column[(column > lower_bound ) & (column < upper_bound)]
```

```python
high_corr_cols = ["emp.var.rate", "euribor3m", "nr.employed"]
```

```python
df1 = df.copy()
df1.columns
```

```python
df1.drop(high_corr_cols, inplace=True, axis=1)
df1.columns
```

```python
df.shape
```

```python
lb = LabelEncoder()
df_encoded = df1.apply(lb.fit_transform)
df_encoded
```

```python
df_encoded["deposit"].value_counts()
```

```python
x = df_encoded.drop("deposit", axis=1)
y = df_encoded["deposit"]
print(x.shape)
print(y.shape)
print(type(x))
print(type(y))
```

```python
x_tr, x_te, y_tr, y_te = train_test_split(x, y, train_size=0.75, random_state=1)
print(x_tr.shape)
print(x_te.shape)
print(y_tr.shape)
print(y_te.shape)
```

```python
def eval_model(y_te, y_pred):
    acc = accuracy_score(y_te, y_pred)
    print("Accuracy_score", acc)
    cm = confusion_matrix(y_te, y_pred)
    print("Confusion Matrix", cm)
    print("Classification Report", classification_report(y_te, y_pred))

def mscore(model):
    train_score = model.score(x_tr, y_tr)
    test_score = model.score(x_te, y_te)
    print("Training Score", train_score)
    print("Test Score", test_score)
```

```python
dt = DecisionTreeClassifier(criterion="gini", max_depth=5, min_samples_split=10)
dt.fit(x_tr, y_tr)
```

```python
mscore(dt)
```

```python
y_pred_dt = dt.predict(x_te)
print(y_pred_dt)
```

```python
eval_model(y_te, y_pred_dt)
```
